```python
from gensim.models import Word2Vec
import pandas as pd
import numpy as np
import re
import os
from nltk import word_tokenize
import pickle
```

```python
# Read the essays and create a dataframe
dataset = pd.concat(pd.read_excel('../data/annotated_xlsx/Attainment_CapitalsPilot_Fall2019.xlsx', sheet_name=None),ignore_index=True,sort=False)
```

```python
dataset = dataset.dropna()
```

```python
dataset.shape
```

```python
essays = dataset['Essay : Why am I here']
```

```python
len(essays)
```

**Clean Essay**

```python
# A helper function to clean the esaay. Remove non-alpha numeric chars, trailng newline, lower case, return the tokenized essay
def clean(essay):
    essay = essay.rstrip('\r\n')
    essay = re.sub('[^\w\s]', ' ', essay) # remove non alpha numeric char
    essay = essay.lower()
    return word_tokenize(essay)
```

**Create Tokens**

```python
tokens = []
for essay in essays:
    if len(essay.split()) !=0:
        tokens.append(clean(essay))
```

```python
len(tokens)
```

```python
tokens
```

**Embeddings**

```python
model = Word2Vec(tokens, min_count=1, size=100, sg=1) #sg=1 -- skipgram
w2v_dict = dict(zip(model.wv.index2word, model.wv.vectors))
print("total words in word2vec dictionary :", len(list(model.wv.vocab)))
```
