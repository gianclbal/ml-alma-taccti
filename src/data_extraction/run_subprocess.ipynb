# Data Extraction Wrapper Notebook

## Objective
The goal of this notebook is to run `extract_iLearn_data.py` through an entire directory. Output is found `output/`.

## Requirements
* Provide the input directory below.

```python
import os
import subprocess
import re

# Specify the input directory containing both .csv (or .xlsx) and .zip files
input_directory = './input/NEW_ALL_DATA'

# List all files in the input directory
all_files = os.listdir(input_directory)

all_files = sorted(all_files)

# Initialize a dictionary to group files by class code
class_files = {}

# Regex pattern to extract the class code
pattern = r'(\w+-\w+-\w+)'

for filename in all_files:
    match = re.search(pattern, filename)
    if match:
        class_code = match.group(1)
    else:
        print("No match found in filename:", filename)
        continue

    if filename.endswith('.csv') or filename.endswith('.xlsx'):
        if class_code not in class_files:
            class_files[class_code] = {'csv': filename, 'zip': []}
        else:
            class_files[class_code]['csv'] = filename
    elif filename.endswith('.zip'):
        if class_code not in class_files:
            class_files[class_code] = {'csv': None, 'zip': [filename]}
        else:
            class_files[class_code]['zip'].append(filename)

class_files
```

```python
import os
import subprocess

# Define the path to the extract_iLearn_data.py script
extract_script = 'extract_iLearn_data.py'
# Iterate through the grouped files and run the script
for class_code, files in class_files.items():
    if files['csv'] and files['zip']:
        input_directory_path = 'input/NEW_ALL_DATA/'
        output_directory_path = 'output/'
        input_file_csv = files['csv']
        input_file_zips = files['zip']

        # Ensure that input_file_zips is a list
        if not isinstance(input_file_zips, list):
            input_file_zips = [input_file_zips]

        # Iterate through ZIP files and run the script for each one
        for input_file_zip in input_file_zips:
            # Construct the command to run the script
            command = [
                'python3',
                extract_script,
                input_directory_path,
                output_directory_path,
                input_file_csv,
                input_file_zip
            ]

            # Run the script
            subprocess.run(command)

    else:
        print(f"Skipping class code {class_code} - Missing CSV or ZIP files")

print("Processing complete.")
```
